```
#escreva  uma função que retorne o valor de um numero
# 1 - o nome da função será o valor_absoluto
# 2 - a função vai receber um numero (inteiro ou flot)
# 3 - a função deve calcular o valor absoluto do numero passado
# 4 - a função deve retornar o valor absoluto
def valor_absoluto(num):
  """Esta função retorna o valor
  absoluto de um número"""

  if num >= 0:
    return num
  else:
    return -num #troca o sinal da função
      
print(valor_absoluto(20))
print(valor_absoluto(-4.5))
```

**Escopo de um código Python**

```
#seja esta função
def minha_funcao(): #Não recebe argumento
  variavel_interna = 300   #variável definida no escopo da função
  print(variavel_interna)

minha_funcao()
```

```
#tentando acessar a variável fora da função
print(variavel_interna)
```

```
#acessando a variável dentro de outra função
def minha_funcao():
  variavel_interna = 300 #variável interna da função
  def minha_funcao_interna(): #função interna da função "minha_funcao"
    print(variavel_interna)
  minha_funcao_interna() #chamado a função interna dentro da função

minha_funcao()
```

**Escopo Global**

```
variavel_externa = 300
def minha_funcao():
  print(variavel_externa)
minha_funcao()

print(variavel_externa)
```

```
#modificando variáveis globais
variavel = 300
def minha_funcao():
  variavel = 200
  print(variavel)

minha_funcao()

print(variavel)
```

```
#utilizando a palavra global
def minha_funcao():
  global variavel_global #utilizar a palavra-chave global para definir uma variavel global no escopo da função
  variavel_global = 300

minha_funcao()

print(variavel_global)
```

```
#trocando o valor da variável global dentro da função altera o valor da variável
x = 300
def minha_funcao():
  global x
  x = 200

minha_funcao()
print(x)
```

```
#função que inverte o valor da string passada
def inverte_string(valor_string):
  resultado=" "
  for caracter in valor_string:
    resultado = caracter +resultado
  return resultado
```

```
#função que retorna verdadeiro ou falso - verifica se um numero x é divsivel pelo numero y
def verifica_divisibildade(x,y):
  x_divisivel_y=False

  if x % y==0:
    x_divisivel_y=True
  return x_divisivel_y
```

```
#construindo o programa via bottom-up
def verificar_divisibilidade(x,y):
    return int(y) % int(x) ==0 # % encontra o resto
```

```
#em um modelo Bottom-UP
#definindo uma função que retorna a lista com os valores da lista que são divisores de x
def elementos_dividiveis(x,lista):
    resultado=[]
    for elemento in lista:
      if verificar_divisibilidade(elemento,x):
        resultado.append(elemento)
    return resultado

```

```
 #função que recebe uma lista de valores e retorna quem são os divisores
 def divisores_em_lista(lista):
    for elemento in lista:
        print(elemento, " é divisivel por", end=" ")
        elementos=elementos_dividiveis(elemento,lista)
        for f in elementos:
            print(f, end=" ")
        print()
```

```
s= input("Entre como um seguencia de valores separados por espaço: ")
lista_valores=[]
for x in s.split():
  lista_valores.append(int(x))
divisores_em_lista(lista_valores)
```

```
#função fatorial - (n! = n*(n-1) * (n-2) *...* 1) - !1=1 !0=1
#5! = 5*4!
#4! = 4*3!
#3! = 3*2!
#2! = 2*1!
#1! = 1
def fatorial(n):
  if n==0:
    return 1

  return n*fatorial(n - 1)
  
print(fatorial(5))
```
