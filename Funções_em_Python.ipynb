**Construindo Funções em Python**
image.png

```
#escreva  uma função que retorne o valor de um numero
# 1 - o nome da função será o valor_absoluto
# 2 - a função vai receber um numero (inteiro ou flot)
# 3 - a função deve calcular o valor absoluto do numero passado
# 4 - a função deve retornar o valor absoluto
def valor_absoluto(num):
  """Esta função retorna o valor
  absoluto de um número"""

  if num >= 0:
    return num
  else:
    return -num #troca o sinal da função
      
print(valor_absoluto(20))
print(valor_absoluto(-4.5))
```

**Escopo de um código Python**

```
#seja esta função
def minha_funcao(): #Não recebe argumento
  variavel_interna = 300   #variável definida no escopo da função
  print(variavel_interna)

minha_funcao()
```

```
#tentando acessar a variável fora da função
print(variavel_interna)
```

```
#acessando a variável dentro de outra função
def minha_funcao():
  variavel_interna = 300 #variável interna da função
  def minha_funcao_interna(): #função interna da função "minha_funcao"
    print(variavel_interna)
  minha_funcao_interna() #chamado a função interna dentro da função

minha_funcao()
```

**Escopo Global**

```
variavel_externa = 300
def minha_funcao():
  print(variavel_externa)
minha_funcao()

print(variavel_externa)
```

```
#modificando variáveis globais
variavel = 300
def minha_funcao():
  variavel = 200
  print(variavel)

minha_funcao()

print(variavel)
```

```
#utilizando a palavra global
def minha_funcao():
  global variavel_global #utilizar a palavra-chave global para definir uma variavel global no escopo da função
  variavel_global = 300

minha_funcao()

print(variavel_global)
```

```
#trocando o valor da variável global dentro da função altera o valor da variável
x = 300
def minha_funcao():
  global x
  x = 200

minha_funcao()
print(x)
```
