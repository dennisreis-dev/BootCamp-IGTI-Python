**O from módulo import atributo**


1.   Importamos apenas o que interessa
2.   Não impporta todo mundo

````
#realizando um import da maneira "mais correia"
from math import factorial
numero=5
print("O factorial de {} é {}". format(numero, factorial(numero) ))
````

from < nome_do_modulo > import* 

````
#utilizando o * para importar tudo
from math import *
numero=6
print("O fatoriak de {} é {}".format(numero,factorial(numero)))
````

**from< nome_do_modulo > import < nome > as < nome_alternativo >**

````
from math import factorial as fatorial #substitui o factorial por fatorial o "as "é um alias
numero=5
print("O fatorial de {} é {}".format(numero,fatorial(numero)))
````
