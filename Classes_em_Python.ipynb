```
#Contrutur de uja classe
#self, serve para referenciar os atributos dentro de um argumento
class carro:
  def __init__(self, cor='vermelho',n_portas=3):
    self.cor_do_carro=cor
    self.n_portas_do_carro=n_portas

  def get_cor_do_carro(self):
    return self.cor_do_carro

  def get_numero_de_portas(self):
    return self.n_portas_do_carro
```

```
#instanciando um objeto do tipo carro
carro_1=carro(cor='azul')
carro_2=carro()
carro_3=carro(cor='prata')
```

```
#mosntrando as cores de cada um dos carros
carro_1.get_cor_do_carro()
```

```
carro_2.get_cor_do_carro()
```

```
carro_3.get_cor_do_carro()
```

```
#Contrutur de uja classe
class carro:
  def __init__(meu_self, cor='vermelho',n_portas=3):
    meu_self.cor_do_carro=cor
    meu_self.n_portas_do_carro=n_portas

  def get_cor_do_carro(meu_self):
    return meu_self.cor_do_carro

  def get_numero_de_portas(meu_self):
    return meu_self.n_portas_do_carro
```

```
#Instanciando um objeto tipo carro
carro_1=carro(cor='azul')
print(carro_2.get_cor_do_carro())
```

**Modificando valores de atributos das classes**

```
carro_1=carro('preto',5)
```

```
print('A cor do carro : ', carro_1.get_cor_do_carro())
print('O carro possui {} portas'.format(carro_1.get_numero_de_portas()))
```

```
#acessando direto os atributos
carro_1.cor_do_carro
```

```
#modificando os atributos
carro_1.cor_do_carro='verde'
```

```
carro_1.cor_do_carro
```

```
#exibindo a cor do carro
print('A nova cor do carro : ',carro_1.get_cor_do_carro())
```
